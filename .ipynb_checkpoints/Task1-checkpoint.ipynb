1) Create dataset as per instruction follow:<br>
Dataset has 2 column as Name and contact no<br>
column name contain unique name<br>
And contact no as follow<br>
first row having 10 digit number<br>
second row having number start with 0<br>
third row having number start with 091<br>
forth row having number start with 87<br>
fift row landline number(6 digit)+ mobile no(10 digit)<br>
sixt row landline number(6 digit)+'  '+ mobile no(10 digit)<br>
seventh row landline number(9 digit)<br>
eight row only landline number(6 digit)<br>


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('abcd.csv')
df['Name'] = df['Name'].apply(lambda x : x.upper())  #All data in string in upper case
print(df)
l = list(df['contact_no'])
ln=[]
mn=[]
for i in l:
    if (len(i)>=6 and len(i)<10):
        l=i
        m=''
        ln.append(l)
        mn.append(m)

    elif len(i)==14 or len(i)==13 or len(i)==12 or len(i)==10:
        if i[-7]==' ':
            l=i
            m=''
            ln.append(l)
            mn.append(m)
        else:
            i=i[-10:]
            l=''
            m=i
            ln.append(l)
            mn.append(m)
            
    elif (len(i)>=16 and len(i)<18):
        rn=i
        m=i[-10:]
        l=rn[0:6]
        ln.append(l)
        mn.append(m)
import re 


def mnValid(s):
    Pattern = re.compile("^[678]\d{9}$")
    return Pattern.match(s) 

s=[]
for i in mn:
    if i=='':
        l='No Mob. Number'
    elif (mnValid(i)):
        l="New Series Number"
    else :
        l="Old Series Number"
    s.append(l)
len(ln)
len(mn)
len(s)
df['landline_No']=ln
df['mobile_No']=mn
df['type_no']=s

df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('Challenge for Anaji.xlsx')
```

```python
m = list(df['Phone'].astype(str))
#print(m)
def clean(l):
    if len(l)==10 or len(l)==11 or len(l)==14 or len(l)==15:
        return l[-10:]
    elif len(l)== 6 or len(l)==8 or len(l)==12:
        return l[-8:]
    elif len(l)>=16:
        lph=[]
        l=l.replace(",",'')
        k=l.split(' ')
        for i in k:
            if (len(i)<10 and len(i)>=6):
                lph.append(i[-8:])
            elif (len(i)>=10):
                lph.append(i[-10:])
        return list(set(lph))
                                
p=[]
for i in m:
    i=clean(i)
    p.append(i)
    
p
```

```python
ph1=[]
ph2=[]
ph3=[]
for i in p:
#     print(type(i))
    if (type(i)==str):
        sp=''
        ph1.append(i)
        ph2.append(sp)
        ph3.append(sp)
        
    else:
        if len(i)==1:
            sp=''
            ph1.append(i[0])
            ph2.append(sp)
            ph3.append(sp)
        elif len(i)==2:
            sp=' '
            ph1.append(i[0])
            ph2.append(i[1])
            ph3.append(sp)
            
        elif len(i)==3:
            sp=' '
            ph1.append(i[0])
            ph2.append(i[1])
            ph3.append(i[2])
            
```

```python
len(ph1)
len(ph2)
len(ph3)
df['Phone1']=ph1
df['Phone2']=ph2
df['Phone3']=ph3
# df['New_Series_Count']
# df['Old_Series_Count']
# df['Landline']

df
```

```python
import numpy as np 
import pandas as pd
import re
data=pd.read_excel('Challenge.xlsx')
print(data)
data['Phone']=data['Phone'].astype(str)
data
#removing duplicates
data = data.groupby(['Name'])['Phone'].apply(', '.join).reset_index()
#making phone column in same format
data['Phone']=data['Phone'].str.replace(', ', ',')
data['Phone']=data['Phone'].str.replace(' ', ',')

#splitting numbers into columns
data[['Phone1','Phone2','Phone3','Phone4']]=data.Phone.apply(lambda x: pd.Series(str(x).split(",")))
#clearing all columns
data.fillna("",inplace=True)

data['Phone1']=data['Phone1'].astype(str).apply(lambda x:"" if len(x)<6 else x)

def clear(s):
    Pattern = re.compile(r'020|022|056')
    match=re.match(Pattern,s)
    if match:
        return s[-8:]
    else:
        return s[-10:]
    
data['Phone1']=data['Phone1'].astype(str).apply(clear)
data['Phone2']=data['Phone2'].astype(str).apply(clear)
data['Phone3']=data['Phone3'].astype(str).apply(clear)
data['Phone4']=data['Phone4'].astype(str).apply(clear)

data.loc[data['Phone1']=="",'Phone1'] = data['Phone2']
data.loc[data['Phone1']==data['Phone2'],'Phone2']=data['Phone3']
data.loc[data['Phone2']==data['Phone3'],'Phone3']=data['Phone4']

data.drop(['Phone','Phone4'],axis=1,inplace=True)

#getting count for new series
def nscount(x):
    pattern=re.compile("^[7-8]")
    match=re.match(pattern,x)
    if match:
        return 1
    else:
        return 0
nl1=data['Phone1'].apply(nscount)
nl2=data['Phone2'].apply(nscount)
nl3=data['Phone3'].apply(nscount)
data['New_Series_Count']=nl1+nl2+nl3

#getting count for old series
def oscount(x):
    pattern=re.compile("^9")
    match=re.match(pattern,x)
    if match:
        return 1
    else:
        return 0
ol1=data['Phone1'].apply(oscount)
ol2=data['Phone2'].apply(oscount)
ol3=data['Phone3'].apply(oscount)
data['Old_Series_Count']=ol1+ol2+ol3

#getting count for landline numbers
def pat(x):
    pattern=re.compile("^[1-6]")
    match=re.match(pattern,x)
    if match:
        return 1
    else:
        return 0
l1=data['Phone1'].apply(pat)
l2=data['Phone2'].apply(pat)
l3=data['Phone3'].apply(pat)
data['Landline']=l1+l2+l3
data
```
